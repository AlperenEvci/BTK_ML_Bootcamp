```python
import matplotlib.pyplot as plt 
plt.plot([1,2,3,4],[1,4,9,16],'r+')
plt.xlabel("Tam Sayilar")
plt.ylabel("Kareler")
plt.title("Bazi tam sayilarin kareleri")
plt.show()
```

```python
import numpy as np 
t = np.arange(0.0,5.0,0.2)
plt.plot(t , t ,"ro--",t ,t**2 ,"bs", t, t**3 ,"g^")
plt.show()
```

```python
data = {'a' : np.arange(50),
        'c' : np.random.randint(0,50,50,),
        'd' : np.random.rand(50)}
data['b'] = data['a'] + (10 * np.random.rand(50))
data['d'] = np.abs(data['d']) * 100

plt.scatter('a','b',c = 'c', s = 'd' , data = data)
plt.xlabel('Values in column a')
plt.ylabel('Values in column b')
plt.show()
```

```python
names = ["Alperen","Osmanalp","Özden","Emine"]
values = ["20","24","49","46"]

plt.figure(figsize = (15,5))

plt.subplot(131)
plt.bar(names,values)

plt.subplot(132)
plt.scatter(names,values)

plt.subplot(133)
plt.plot(names,values)

plt.suptitle("Graphs")
```
