```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 


import warnings
warnings.filterwarnings("ignore")

from plotly.offline import init_notebook_mode , iplot , plot
import plotly as py 
init_notebook_mode(connected=True)
import plotly.graph_objs as go 
import plotly.io as pio 
import plotly.figure_factory as ff 
```

```python
dt = pd.read_csv("University_Rankings_20220204.csv")
dt.head()
```

```python
dt.columns
```

```python
df = dt.iloc[:100,:]
df.head()
```

```python
#ploty stilleri
pio.templates.default = "simple_white"
```

```python
line1 = go.Scatter(x = df.world_rank , 
                   y = df.citations ,
                   mode = "markers",
                   name = "Citations" , 
                   marker= dict(color = "rgba(78,78,250,0.85)"),
                   text  = df.university_name )


line2 = go.Scatter(x = df.world_rank , 
                   y = df.teaching ,
                   mode = "markers",
                   name = "Eğitim-Öğretim Faaliyetleri" , 
                   marker= dict(color = "rgba(202,43,15,0.85)"),
                   text  = df.university_name )

dt = [line1,line2]
res = dict(title = "Dünta Siralamasinda ilk 100 uninin citations ve eğitim-öğretim faaliyetleri",xaxis = dict(title = "Word Ranking",ticklen = 5, zeroline = False))
fig = dict(data = dt ,layout = res )
plot(fig,filename= "1_cizgi_alinti_ve_ogretim_puanlari.html")
```

```python
dt2011 = df[df.year == 2011].iloc[:5:]

line1 = go.Bar(x = dt2011.university_name , 
                   y = dt2011.citations ,
                   name = "Citations" , 
                   marker= dict(color = "rgba(255,127,40,0.5)",
                   line = dict(color = "rgb(0,0,0)",width = 1.5)),
                   text  = dt2011.country )

line2 = go.Bar(x = dt2011.university_name , 
                   y = dt2011.teaching ,
                   name = "Eğitim-Öğretim Faaliyetleri" , 
                   marker= dict(color = "rgba(64,10,128,0.5)",
                   line = dict(color = "rgb(0,0,0)",width = 1.5)),
                   text  = dt2011.country )

dt_= [line1,line2]
res = go.Layout(barmode = "group")
fig = go.Figure(data = dt_, layout= res)
plot(fig,filename = "2_cubuk_alinti_ve_ogretim_puanlari.html")
```

```python
dt_a = df[df.year == 2011].iloc[:8:]

pie1 = dt_a.num_students 
pie1_list = [float(each.replace(",",".")) for each in dt_a.num_students]
labels = dt_a.university_name

line = go.Pie(labels = labels,
              values = pie1_list,
              hoverinfo= "label+value+percent",
              textinfo= "value+percent",
              textfont= dict(size = 8),
              rotation= 180,
              hole = 0.3,
              marker = dict(line =  dict(color = "#000000" ,width = 1)))


dt_ = [line]
res = dict(title = "2011 yili - ilk 8 üniversitenin öğrenci sayilari ve oranlari",legend = dict(orientation = "h"))
fig = dict(data = dt_ , layout = res)
plot(fig, filename = "3_dairesel_ogrenci_sayisi.html")

```
