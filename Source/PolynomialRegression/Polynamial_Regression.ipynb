```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression # for linear regression
from sklearn.preprocessing import PolynomialFeatures # for polynomial regression

# Importing the dataset
dataset = pd.read_csv('polynomial_regression.csv' , sep= ';')
dataset.head()
```

```python
#sklearn requires the input data to be in 2D array numpyarray
y =  dataset.araba_max_hiz.values.reshape(-1,1) # dependent variable to convert 2D array
x = dataset.araba_fiyat.values.reshape(-1,1) # independent variable to convert 2D array
```

```python
#data visualization
plt.scatter(x,y)
plt.xlabel('araba_fiyat')
plt.ylabel('araba_max_hiz')
plt.grid(True)
plt.show()
```

```python
#linear regression model fitting
lr = LinearRegression()
#fitting the model
lr.fit(x,y)
```

```python
#%% tahmin
y_head = lr.predict(x)

#plotting the result
plt.scatter(x,y)
plt.plot(x,y_head,color = 'red')
plt.xlabel('araba_fiyat')
plt.ylabel('araba_max_hiz')
plt.grid(True)
plt.show()

car_price = 10000
print("10M lik araba hiz tahmini",lr.predict([[car_price]]))
```

```python
#polynomial regression  = y = b0 + b1*x + b2*x^2 + b3*x^3 + ... + bn*x^n
#polinomsal özellikler 
polynomial_regression = PolynomialFeatures(degree = 4) # degree = n 4. dereceden polinom
x_polynomial = polynomial_regression.fit_transform(x)
x_polynomial

#çıktılarda 1, x, x^2, x^3, x^4 şeklinde sütunlar oluştu
```

```python
#polynomial regression model fitting 
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)
```

```python
#polynomial regression visualization
y_head2 = linear_regression2.predict(x_polynomial)
plt.scatter(x,y)
plt.plot(x,y_head,color = 'red', label = 'linear')
plt.plot(x,y_head2,color = 'green' , label = 'polynomial')
plt.legend()
plt.xlabel('araba_fiyat')
plt.ylabel('araba_max_hiz')
plt.title('Araba Fiyatına Göre Hız Tahmini (Polynomial Regression)')
plt.grid(True)
plt.show()
```
