```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#sklearn imports
from sklearn.linear_model import LinearRegression   #for linear regression
from sklearn.preprocessing import PolynomialFeatures  #for polynomial regression
from sklearn.metrics import r2_score  #for r2 score

#read the data
data = pd.read_csv('linear_regression.csv',sep=';')
data.head()
```

```python
#data visualization
plt.scatter(data.deneyim,data.maas)
plt.xlabel('deneyim')
plt.ylabel('maas')
plt.title('deneyim-maas ilişkisi')
plt.grid(True)
plt.show()
```

```python
#data içerisinde bulunan deneyim ve maas verilerini x ve y olarak atama 
x = data.deneyim.values.reshape(-1,1)
y = data.maas.values.reshape(-1,1)

#model oluşturma
linear_reg = LinearRegression()
linear_reg.fit(x,y)
```

```python
#polynomial regression model eğitimi
polynomial_reg = PolynomialFeatures(degree=2) #2. dereceden polinom
x_polynomial = polynomial_reg.fit_transform(x)

#polinomsal model oluşturmak için 
poly_reg = LinearRegression()
poly_reg.fit(x_polynomial,y)
```

```python
#test
y_pred_linear = linear_reg.predict(x) #linear regression ile tahmin
y_pred_polynomial = poly_reg.predict(x_polynomial) #polynomial regression ile tahmin

plt.scatter(data.deneyim,data.maas)
plt.plot(x,y_pred_linear,color='red',label='linear') #linear regression ile tahmin
plt.plot(x,y_pred_polynomial,color='green',label='polynomial')  #polynomial regression ile tahmin
plt.xlabel('deneyim')
plt.ylabel('maas')
plt.title('deneyim-maas ilişkisi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#r2 score
print('linear r2 score:',r2_score(y,y_pred_linear))
print('polynomial r2 score:',r2_score(y,y_pred_polynomial))
```
